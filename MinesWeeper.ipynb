Pravimo matricu koja prikazuje okolne bombe

```python
m, n = 9,9
broj_bombi = [[0] * n for _ in range(m)]
for v in range(m):
    for k in range(n):
        broj_bombi[v][k] = 0
        for dv in range(-1, 2):
            for dk in range(-1, 2):
                vv, kk = v + dv, k + dk
                if (vv != v or kk != k) and (0 <= vv < m) and (0 <= kk < n) and matrica1[vv][kk]:
                    broj_bombi[v][k] += 1

# ispisujemo rezultat
for v in range(m):
    print(" ".join(map(str, broj_bombi[v])))

```

```python
def uncover_field(i, j, minefield):
    print(i, j)
    if 0 <= i < len(minefield) and 0 <= j < len(minefield[0]):
        if minefield[i][j] == 0:
            minefield[i][j] = -10
            for y in range(-1, 2):
                for x in range(-1, 2):
                    uncover_field(i + y, j + x, minefield)
        elif 0 < minefield[i][j] < 9:
            minefield[i][j] *= -1

def draw_minefield(minefield):

    print('''    0   1   2   3   4   5   6   7   8
  -------------------------------------------''')

    for i in range(len(minefield)):
        print(i,"|", end="")
        for j in range(len(minefield[0])):

            if minefield[i][j] >= 0:  # Covered field
                print(" # ", end=" ")
            elif minefield[i][j] == -10:  # Uncovered empty field
                print(" 0 ", end=" ")
            elif minefield[i][j] == -9:  # Uncovered mine
                print(" X ", end=" ")
            else:  # Uncovered number
                print(" " + str(-minefield[i][j])[0] + " ", end=" ")
        print()
    print()

```

```python
flag=True

draw_minefield(broj_bombi)
while flag:

  i1 = int(input('Unesi prvu koordinatu (0-8): '))
  j1 = int(input('Unesi drugu koordinatu (0-8): '))

  if not (0 <= i1 < redovi and 0 <= j1 < kolone):
    print("Koordinate moraju biti u opsegu od 0 do 8.")
    break

  if matrica1[i1][j1] == 1:
    print("Izgubio si D:")
    flag = False
  else:
    uncover_field(i1,j1,broj_bombi)
    draw_minefield(broj_bombi)
```
